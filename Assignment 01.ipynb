#median

```python
median_satis = data['Pregnancies'].median()
```

```python
median_satis
```

```python
data['Pregnancies'] = data['Pregnancies'].fillna(median_satis)
```

```python
median_satis = data['SkinThickness'].median()
```

```python
data['SkinThickness'] = data['SkinThickness'].fillna(median_satis)
```

```python
median_satis = data['Insulin'].median()
```

```python
data['Insulin'] = data['Insulin'].fillna(median_satis)
```

#min max scaler

```python
x = data.drop('Outcome',axis=1)
y = data['Outcome']
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
x.loc[:,['Glucose', 'BloodPressure','SkinThickness','Insulin','BMI']] = min_max.fit_transform(x[['Glucose', 'BloodPressure', 'SkinThickness','Insulin','BMI',]])
```

```python
x.head()
```

#model training

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train,y_train) #error coming as total no iterations reached limit
```

```python
y_pred = log_reg.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python

from sklearn.metrics import confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
