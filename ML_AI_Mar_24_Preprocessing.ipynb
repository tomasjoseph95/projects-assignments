```python
data['number_of_projects'] = data['number_of_projects'].fillna('four')
```

```python
plt.hist(data['satisfaction'])
plt.title('Distribution of Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('frequency')
plt.show()
```

```python
median_satis = data['satisfaction'].median()
```

```python
median_satis
```

```python
data['satisfaction'] = data['satisfaction'].fillna(median_satis)
```

```python
plt.hist(data['evaluation'])
plt.title('Distribution of evaluation')
plt.xlabel('Evaluation')
plt.ylabel('frequency')
plt.show()
```

```python
data['evaluation'] = data['evaluation'].fillna(data['evaluation'].median())
```

```python
data.isna().sum()
```

```python
plt.hist(data['average_montly_hours'])
plt.title("distribution of average_montly_hours")
plt.xlabel("average_montly_hours")
plt.ylabel("frequency")
plt.show()
```

```python
mean_amh = data['average_montly_hours'].mean()
data['average_montly_hours'] = data['average_montly_hours'].fillna(mean_amh)
```
